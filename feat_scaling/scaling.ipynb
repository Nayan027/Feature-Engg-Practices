```python
# Get columns where dtype is not object
numeric_columns = data.columns[data.dtypes != 'object']

print(numeric_columns)
```

```python
# Keep only numeric columns in a new DataFrame
numeric_col_df = data.select_dtypes(exclude=['object'])

numeric_col_df.head()
```

```python
y = numeric_col_df[["overall_performance_score"]]   # double brackets makes it a DataFrame
y.head()
```

```python
X = numeric_col_df.drop(columns=["overall_performance_score"])
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
X_train.head(2)
```

```python
y_train.head(2)
```

```python
from sklearn.preprocessing import MinMaxScaler           # Min-Max Scaling (values scaled to [0,1])

minmax_scaler = MinMaxScaler()
X_train_minmax = minmax_scaler.fit_transform(X_train)
X_test_minmax  = minmax_scaler.transform(X_test)
```

```python
import pandas as pd                                      # back in form of DataFrames 

X_train_minmax = pd.DataFrame(X_train_minmax, columns=X_train.columns, index=X_train.index)
X_test_minmax  = pd.DataFrame(X_test_minmax, columns=X_test.columns, index=X_test.index)

X_train_minmax.head(2)
```

```python
from sklearn.preprocessing import  StandardScaler                # Mean Normalization (Standardization: mean=0, std=1)

standard_scaler = StandardScaler()
X_train_standard = standard_scaler.fit_transform(X_train)
X_test_standard  = standard_scaler.transform(X_test)
```

```python
X_train_standard = pd.DataFrame(X_train_standard, columns=X_train.columns, index=X_train.index)
X_test_standard  = pd.DataFrame(X_test_standard, columns=X_test.columns, index=X_test.index)

X_train_standard.head(2)
```

```python
from sklearn.preprocessing import MaxAbsScaler                     # Max Absolute Scaling (scales by max abs value)

maxabs_scaler = MaxAbsScaler()
X_train_maxabs = maxabs_scaler.fit_transform(X_train)
X_test_maxabs  = maxabs_scaler.transform(X_test)
```

```python
X_train_maxabs = pd.DataFrame(X_train_maxabs, columns=X_train.columns, index=X_train.index)
X_test_maxabs  = pd.DataFrame(X_test_maxabs, columns=X_test.columns, index=X_test.index)

X_train_maxabs.head(2)
```

```python
from sklearn.preprocessing import RobustScaler             # Robust Scaling (uses median & IQR)

robust_scaler = RobustScaler()
X_train_robust = robust_scaler.fit_transform(X_train)
X_test_robust  = robust_scaler.transform(X_test)
```

```python
X_train_robust = pd.DataFrame(X_train_robust, columns=X_train.columns, index=X_train.index)
X_test_robust  = pd.DataFrame(X_test_robust, columns=X_test.columns, index=X_test.index)

X_train_robust.head(2)
```
